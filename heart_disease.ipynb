### Attributes Information

**age:** age

**sex:** sex

**cp:** chest pain type (4 values)

**trestbps:** resting blood pressure

**chol:** serum cholestoral in mg/dl

**fbs:** fasting blood sugar > 120 mg/dl

**resrecg:** resting electrocardiographic results (values 0,1,2)

**thalach:** maximum heart rate achieved

**exang:** exercise induced angina

**oldpeak:** ST depression induced by exercise relative to rest

**slope:** the slope of the peak exercise ST segment

**ca:** number of major vessels (0-3) colored by flourosopy

**thal:** 0 = normal; 1 = fixed defect; 2 = reversable defect

**target:** affected

```python
# Assigning values to X and y
X = df.drop('target', axis=1)
y = df['target']
```

```python
# Assigning values to variables and splitting dataset for training
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Importing our model and fitting it
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression().fit(X_train, y_train)
```

```python
# Getting predictions from our trained model
y_pred = LR.predict(X_test)
```

```python
# Checking accuracy and confusion matrix of our trained model

from sklearn.metrics import accuracy_score, confusion_matrix

print(accuracy_score(y_test, y_pred))
c_m = confusion_matrix(y_test, y_pred)
print(c_m)
```

```python
sns.heatmap(data=c_m, annot=True)
plt.show()
```

```python
# Importing model and checking its accuracy

from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier().fit(X_train, y_train)

y_pred = RF.predict(X_test)

print(accuracy_score(y_test, y_pred))
```
